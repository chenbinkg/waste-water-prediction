```python
def prep_rainfall(df):
    '''
    preprocess rainfall data
    '''
    df['hour'] = df['Time'].dt.hour
    df['wday'] = df['Time'].dt.dayofweek
    df['month'] = df['Time'].dt.month
    df['mday'] = df['Time'].dt.days_in_month
    df["doy"] = df['Time'].dt.dayofyear
    df['Rainlag1'] = df['Rainfall'].shift(1)
    df['Rainlag2'] = df['Rainfall'].shift(2)
    df['Rainlag1'] = df['Rainlag1'].fillna(method="bfill")
    df['Rainlag2'] = df['Rainlag2'].fillna(method="bfill")
    df['Rain_L3HR'] = df.loc[:,'Rainfall'].rolling(window=3).sum()
    df['Rain_L6HR'] = df.loc[:,'Rainfall'].rolling(window=6).sum()
    df['Rain_L12HR'] = df.loc[:,'Rainfall'].rolling(window=12).sum()
    df['Rain_L24HR'] = df.loc[:,'Rainfall'].rolling(window=24).sum()
    df['Rain_L48HR'] = df.loc[:,'Rainfall'].rolling(window=48).sum()
    df['Rain_L3HR'] = df['Rain_L3HR'].fillna(method="bfill")
    df['Rain_L6HR'] = df['Rain_L6HR'].fillna(method="bfill")
    df['Rain_L12HR'] = df['Rain_L12HR'].fillna(method="bfill")
    df['Rain_L24HR'] = df['Rain_L24HR'].fillna(method="bfill")
    df['Rain_L48HR'] = df['Rain_L48HR'].fillna(method="bfill")
    
    df['sin_hour'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['cos_hour'] = np.cos(2 * np.pi * df['hour'] / 24)
    
    # Typical value between 0.8 and 0.95
    k_api = 0.85
    
    # Initialize API and Soil Moisture as columns
    df['API'] = 0
    df['SoilMoisture'] = 0
    
    # Calculate Antecedent Precipitation Index (API)
    for i in range(1, len(df)):
        df.at[i, 'API'] = k_api * df.at[i - 1, 'API'] + df.at[i - 1, 'Rainfall']
    
    # Soil Moisture Index (very simplified: saturates at some threshold)
    # Arbitrary max bucket size (mm)
    max_storage = 100
    # Moisture decay rate
    soil_decay = 0.95
    # Initial condition
    df.at[0, 'SoilMoisture'] = min(max_storage, df.at[0, 'Rainfall'])
    
    for i in range(1, len(df)):
        df.at[i, 'SoilMoisture'] = min(
            max_storage,
            df.at[i - 1, 'SoilMoisture'] * soil_decay + df.at[i - 1, 'Rainfall']
        )

    return df
```

```python
import s3fs
import boto3

bucket_name = 'niwa-water-demand-modelling'
fs = s3fs.S3FileSystem(anon=False)

rf_dict = {
    'PCD Upper Hutt wet weather flow': 'Pinehaven (mm/hr)',
    'PCD Stokes Valley wet weather flow': 'Birch Lane (mm/hr)',
    'PCD Lower Hutt wet weather flow': 'Birch Lane (mm/hr)',
    'SilverstreamAllInflow': 'Pinehaven (mm/hr)',
    'UpperHutt': 'Pinehaven (mm/hr)'
}

# populationUpper = 48823
# populationStokes = 6259
# populationLower = 107917
# populationTotal = 162999

population_dict = {
    'SilverstreamAllInflow': 55082,
    'UpperHutt': 48823
}

# take hydroclimate observation
pcd_file_name = "HVWW_hydroclimateV8_calib_lhWWF_recess14_16_18.csv"
header = pd.read_csv(pcd_file_name, nrows=12)
header = ['hour', 'day_num'] + header.iloc[:,0].values[-7:].tolist()
obs_targets = [e for e in header if 'wet weather' in e]
df_obs = pd.read_csv(pcd_file_name, skiprows=13, names=header)
df_obs = df_obs.reset_index(drop=True)
columns_sel = ['hour', 'day_num'] + [e for e in header if 'wet weather' in e]
df_obs = df_obs[columns_sel]
start_date = pd.to_datetime('2008-01-01 00:00:00')
end_date = pd.to_datetime('2017-12-31 23:00:00')
dt_idx = pd.date_range(start_date, end_date, freq='h')
dt_idx = dt_idx.strftime(date_format='%Y-%m-%d %H:%M:%S')
dt_idx_removed_leapyear = [e for e in dt_idx if '-02-29' not in e]
df_obs['Time'] = dt_idx_removed_leapyear
df_obs = df_obs.set_index(pd.to_datetime(df_obs["Time"], format="%Y-%m-%d %H:%M:%S"))

# read in rainfall
df_rf = pd.read_csv("HuttModelRainfallPinehavenBirch2008-2017.csv")
df_rf = df_rf.rename(columns={'P_DATETIME': 'Time'})
df_rf = df_rf.set_index(pd.to_datetime(df_rf["Time"], format="%d/%m/%Y %H:%M:%S"))
# divide by 12 first
df['Birch Lane (mm/hr)'] = df['Birch Lane (mm/hr)']/12
df['Pinehaven (mm/hr)'] = df['Pinehaven (mm/hr)']/12
df_rf = df_rf.resample('H').sum(['Birch Lane (mm/hr)', 'Pinehaven (mm/hr)'])

df_obs_cur = pd.read_csv("WWF_hourly_flow_CUR_2023.csv", index_col=[0])
df_obs_cur['UpperHutt'] = df_obs_cur['SilverstreamAllInflow'] - df_obs_cur['StokesValley to Silverstream']
df_obs_cur = df_obs_cur.set_index(pd.to_datetime(df_obs_cur["Time"], format="%Y-%m-%d %H:%M:%S"))
df_obs_all = df_obs[obs_targets].join(df_obs_cur[['UpperHutt', 'SilverstreamAllInflow']])
df_obs_all = df_obs_all.join(df_rf)
df_obs_all.to_csv("obs_all.csv")

# read in rainfall
df_temp = pd.read_csv("Temperature.csv")
df_temp['Date'] = pd.to_datetime(df_temp['Date'], format="%d/%m/%Y %H:%M")
df_temp = df_temp.set_index('Date')

for obs_target in obs_targets:
    rainfall_var = rf_dict[obs_target]
    df_rf_1 = df_rf.rename(columns={rainfall_var: 'Rainfall'})
    if obs_target in population_dict.keys():
        population = population_dict[obs_target]
    else:
        population = None
    print(f"{obs_target} - rainfall: {rainfall_var}, population: {population}")
    df_rf_1 = prep_rainfall(df_rf_1[['Rainfall']].reset_index())
    df = df_obs[[obs_target]].join(df_rf_1.set_index('Time')).join(df_temp['Dry bulb degC'])
    df = df.reset_index()
    if population:
        df[obs_target] = abs(df[obs_target])*3600/population
    
    prefix = obs_target.replace(" ", "_")
        
    if rainfall_var == 'Birch Lane (mm/hr)':
        file_name = f"{prefix}_birch_lane.csv"
        df.to_csv(file_name, index=False)
    else:
        file_name = f"{prefix}_pine_haven.csv"
        df.to_csv(file_name, index=False)
    
    s3_path = f's3://{bucket_name}/WWL/{file_name}'
    with fs.open(s3_path, 'w') as f:
        df.to_csv(f, index=False)
```
